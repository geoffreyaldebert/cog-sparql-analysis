```python
import pandas as pd
import json
```

On récupère les fichiers générés à l'étape 0

```python
with open('coms.json') as json_file:
    coms = json.load(json_file)
with open('evts.json') as json_file:
    evts = json.load(json_file)
```

```python
dfcom = pd.DataFrame(coms)
dfevt = pd.DataFrame(evts)
```

```python
dfevt
```

Enrichissement du dataframe événement en ajoutant le détail de chaque commune ajoutée ou supprimée dans deux nouvelles colonnes

```python
def getComs(coms):
    arr = []
    for com in coms:
        mydict = dfcom[dfcom['id'] == com].to_dict(orient = 'records')
        if(len(mydict) > 0):
            arr.append(dfcom[dfcom['id'] == com].to_dict(orient = 'records')[0])
    return arr
```

```python
dfevt['creationObj'] = dfevt['creation'].apply(lambda x: getComs(x))
```

```python
dfevt['suppressionObj'] = dfevt['suppression'].apply(lambda x: getComs(x))
```

Sauvegarde dans un fichier (le traitement est long donc inutile de le rejouer à chaque fois)

```python
import json
with open('evts-detail.json', 'w') as f:
    json.dump(dfevt.to_dict(orient="records"), f, indent=4)
```
